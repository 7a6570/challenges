```python
from itertools import permutations
import struct
```

```python
ror = lambda val, r_bits, max_bits: \
    ((val & (2**max_bits-1)) >> r_bits%max_bits) | \
    (val << (max_bits-(r_bits%max_bits)) & (2**max_bits-1))
```

```python
def get_uppercase_alphabet():
    
    char=65
    alphabet="A"
    for i in range(25):
        
        char=ord(alphabet[-1])
        next_char=chr(char+1)
        alphabet+=next_char        
        
    return alphabet
```

```python
l=permutations(get_uppercase_alphabet(),4)
```

```python
def transform_key(ebp, esi=0, edx=0):
    """
        add edx, 4
        xor ebp, esi
        ror ebp, 3
        sub ebp, edx
        xor esi, ebp
        
        :param ebp: key, integer
    """
    
    edx=edx+4
    ebp = ebp ^ esi
    ebp=ror(ebp,3,32)
    ebp=ebp-edx
    esi=esi ^ ebp
    
    return esi
```

```python
struct.pack(">L", transform_key(struct.unpack(">L", "GOOG".encode('ascii'))[0]))
```

```python
def decrypt(key):
    
    key_int=struct.unpack(">L", key.encode('ascii'))[0] 
    key_int=transform_key(key_int)
    cipher_int=struct.unpack("<L", bytes.fromhex("567d8dde"))[0]
    
    key_bytes = struct.pack(">L", key_int).hex()
    cipher_bytes = struct.pack(">L", cipher_int).hex()
    
    #print(f"esi bytes: {key_bytes}")
    #print(f"esp +4 bytes: {cipher_bytes}")
    
    plaintext_bytes = cipher_int ^ key_int  
    
    return struct.pack("<L", plaintext_bytes)
```
